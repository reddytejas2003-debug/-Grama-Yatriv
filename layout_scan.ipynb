```python
import os
import xml.etree.ElementTree as ET
root = os.path.join('app','src','main','res')
errors = []
for dirpath, dirnames, filenames in os.walk(root):
    if os.path.basename(dirpath).startswith('layout'):
        for fn in filenames:
            if fn.endswith('.xml'):
                path = os.path.join(dirpath, fn)
                try:
                    tree = ET.parse(path)
                    root_el = tree.getroot()
                except Exception as e:
                    errors.append((path, 'PARSE_ERROR', str(e)))
                    continue
                for elem in root_el.iter():
                    tag = elem.tag
                    if isinstance(tag, str) and tag.startswith('{'):
                        tag = tag.split('}', 1)[1]
                    if tag in ['merge', 'request', 'include', 'fragment']:
                        continue
                    attrs = elem.attrib
                    if 'android:layout_width' not in attrs or 'android:layout_height' not in attrs:
                        errors.append((path, ET.tostring(elem, encoding='unicode').strip().split('\n')[0], 'missing width or height'))
print('COUNT', len(errors))
for p, tag, reason in errors:
    print(p, reason, tag)
```
